```python
# [0,1,2] total value of 3
x = 2
# [o,1,2] total value of 3
y = 2
#[0,1] total value of 2
z = 1

# Find total number of coordinates
#Total Points = (x+1) x (y+1) x (z+1)
# 3 X 3 X 2 = 18 and so the rest

# sum the total n
coordinates = [[i, j, k] for i in range(x+1) for j in range(y+1) for k in range(z+1)]

# Print the result
print("Total number of coordinates:", len(coordinates))
print("Coordinates:", coordinates)
```

## Added Condition

The Hacker Rank wants the sum of i+j+k value is not equal to n value, this code used to filter conditions of the 3D cubic

for examples:
lets take examples
x = 2
y = 2
z = 1
n = 2

Case
- [0,0,0] = 0 then includes as 1 coordinates
- [0,1,0] = 1 then includes as 1 coordinates
- [0,2,0] = 2 then excludes, because we are filter if sum of the value not same as n(2)
- etc

# Solve

```python
x = 2
y = 2
z = 1
n = 2

#sum total n
coordinates = [[i,j,k] for i in range(x+1) for j in range(y+1) for k in range(z+1) if i+j+k != n]
print("Total number of coordinates:", len(coordinates))
print("Coordinates", coordinates)
```
